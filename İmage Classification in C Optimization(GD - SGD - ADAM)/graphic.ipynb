```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

# Ağırlık dosyalarını okuma fonksiyonu
def read_weights(file_path):
    weights = []
    with open(file_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            if line.startswith('w['):
                values = list(map(float, line.split(':')[1].strip().split()))
                weights.append(values)
    return np.array(weights)

# Dosya yolları (1. başlangıç seti için tüm modeller)
file_paths = {
    "Adam1": "weights_gd_adam_1.txt",
    "Adam2": "weights_gd_adam_2.txt",
    "Adam3": "weights_gd_adam_3.txt",
    "Adam4": "weights_gd_adam_4.txt",
    "Adam5": "weights_gd_adam_5.txt"
    
}

# Modellerin ağırlıklarını okuma
trajectories = {}
for model_name, file_path in file_paths.items():
    weights = read_weights(file_path)
    trajectories[model_name] = weights[:100]  # İlk 100 epoch ile sınırlandır

# Tüm ağırlıkları birleştirme
all_weights = np.vstack([trajectories[model_name] for model_name in trajectories])

# PCA ile boyut indirgeme (t-SNE öncesi)
pca = PCA(n_components=4)
reduced_data = pca.fit_transform(all_weights)

# t-SNE ile boyut indirgeme
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(reduced_data)

# t-SNE sonuçlarını ayırma
start_idx = 0
tsne_trajectories = {}
for model_name in trajectories:
    num_points = trajectories[model_name].shape[0]
    tsne_trajectories[model_name] = tsne_result[start_idx:start_idx + num_points]
    start_idx += num_points

# Görselleştirme
plt.figure(figsize=(12, 8))
colors = {"Adam1": "r", "Adam2": "g", "Adam3": "b","Adam4": "y","Adam5": "m"}
for model_name, tsne_points in tsne_trajectories.items():
    # Başlangıç noktası
    plt.scatter(tsne_points[0, 0], tsne_points[0, 1], color=colors[model_name], edgecolor='k', s=100, label=f"{model_name} Start")
    # Bitiş noktası
    plt.scatter(tsne_points[-1, 0], tsne_points[-1, 1], color=colors[model_name], edgecolor='k', s=100, marker='X', label=f"{model_name} End")

plt.title("Aynı Başlangıç Noktasından Farklı Modellerin Sapmaları (t-SNE)")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.legend()
plt.grid()
plt.show()


# Ağırlık dosyalarını okuma fonksiyonu
def read_weights(file_path):
    weights = []
    with open(file_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            if line.startswith('w['):
                values = list(map(float, line.split(':')[1].strip().split()))
                weights.append(values)
    return np.array(weights)

# Dosya yolları (1. başlangıç seti için tüm modeller)
file_paths = {
    "GD1": "weights_gd_1.txt",
    "GD2": "weights_gd_2.txt",
    "GD3": "weights_gd_3.txt",
    "GD4": "weights_gd_4.txt",
    "GD5": "weights_gd_5.txt"
    
}

# Modellerin ağırlıklarını okuma
trajectories = {}
for model_name, file_path in file_paths.items():
    weights = read_weights(file_path)
    trajectories[model_name] = weights[:100]  # İlk 100 epoch ile sınırlandır

# Tüm ağırlıkları birleştirme
all_weights = np.vstack([trajectories[model_name] for model_name in trajectories])

# PCA ile boyut indirgeme (t-SNE öncesi)
pca = PCA(n_components=4)
reduced_data = pca.fit_transform(all_weights)

# t-SNE ile boyut indirgeme
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(reduced_data)

# t-SNE sonuçlarını ayırma
start_idx = 0
tsne_trajectories = {}
for model_name in trajectories:
    num_points = trajectories[model_name].shape[0]
    tsne_trajectories[model_name] = tsne_result[start_idx:start_idx + num_points]
    start_idx += num_points

# Görselleştirme
plt.figure(figsize=(12, 8))
colors = {"GD1": "r", "GD2": "g", "GD3": "b","GD4": "y","GD5": "m"}
for model_name, tsne_points in tsne_trajectories.items():
    # Başlangıç noktası
    plt.scatter(tsne_points[0, 0], tsne_points[0, 1], color=colors[model_name], edgecolor='k', s=100, label=f"{model_name} Start")
    # Bitiş noktası
    plt.scatter(tsne_points[-1, 0], tsne_points[-1, 1], color=colors[model_name], edgecolor='k', s=100, marker='X', label=f"{model_name} End")

plt.title("Aynı Başlangıç Noktasından Farklı Modellerin Sapmaları (t-SNE)")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.legend()
plt.grid()
plt.show()

```

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
def load_weights(file_path):
    """
    Belirtilen dosyadan ağırlık verilerini yükler.
    """
    weights = []
    with open(file_path, 'r') as f:
        current_epoch = -1
        epoch_weights = []
        for line in f:
            if line.startswith("Epoch"):
                if current_epoch != -1:
                    weights.append(epoch_weights)
                epoch_weights = []
                current_epoch += 1
            else:
                epoch_weights.extend([float(x) for x in line.strip().split()[1:]])
        weights.append(epoch_weights)
    return np.array(weights)

def plot_tsne_trajectories(file_paths, labels, colors, markers, title):
    """
    t-SNE ile ağırlık trajektoryalarını indirger ve tek bir grafikte gösterir.
    """
    plt.figure(figsize=(10, 8))

    for file_path, label, color, marker in zip(file_paths, labels, colors, markers):
        weights = load_weights(file_path)
        tsne = TSNE(n_components=2, random_state=42, perplexity=30)
        weights_2d = tsne.fit_transform(weights)
        plt.plot(weights_2d[:, 0], weights_2d[:, 1], marker=marker, label=label, color=color)
    
    plt.title(title)
    plt.xlabel("t-SNE Dimension 1")
    plt.ylabel("t-SNE Dimension 2")
    plt.legend()
    plt.grid(True)
    plt.show()

# 5 farklı weights_gd.txt dosyasını belirtin, her birini GD, SGD ve Adam için ekleyin
file_paths = [
    "weights_sgd_1.txt", "weights_sgd_2.txt", "weights_sgd_3.txt", "weights_sgd_4.txt", "weights_sgd_5.txt"
]

# Her dosya için bir etiket ve renk tanımlayın
labels = [
    "SGD Initial w1", "SGD Initial w2", "SGD Initial w3", "SGD Initial w4", "GD Initial w5"
]

colors = [
    "red", "blue", "green", "purple", "orange"
]

# Her optimizasyon yöntemi için farklı markerlar tanımlayın
markers = [
    "o", "o", "o", "o", "o"
]

# Grafiği oluştur
plot_tsne_trajectories(
    file_paths=file_paths,
    labels=labels,
    colors=colors,
    markers=markers,
    title="t-SNE Trajectories of Weight Updates"
)
```
